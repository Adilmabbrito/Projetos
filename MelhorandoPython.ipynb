Escreva um programa que encontre todos os numeros divisiveis por 7, mas que não sejam multiplos de 5, entre 2000 e 3000(ambos incluidos). Os numeros obtidos devem ser impressos em sequencia separada por vírgulas em uma única linha.

```python
numeros_encontrados = []
for numero in range(2000,3001):
  if numero %7 ==0 and numero % 5 !=0:
    numeros_encontrados.append(numero)
    sequencia = ','.join(map(str, numeros_encontrados))
    print(sequencia)
```

Escreva um programa que possa calcular o fatorial de um dado número. Os resultados devem ser impressos em uma sequência separada por vírgulas em uma única linha. Suponha que a seguinte entrada seja fornecida ao programa:8. Então a saída deve ser: 40320

```python
def calcular_fatorial(numero):
  if numero ==0 or numero == 1:
    return 1
  else:
    return numero * calcular_fatorial(numero -1)

numero = int(input("Digite um número para calcular o fatorial: "))

resultado = calcular_fatorial(numero)

print(resultado)
```

Escreva um programa que aceite uma sequência de números separados por vírgulas do console e gere uma lista e uma tupla que contenha todos os números. Suponha que a seguinte entrada seja fornecida ao programa: 34,67,55,33,12,98. Então a saída deve ser: ['34', '67', '55', '33', '12', '98'] ('34','67', '55', '33', '12', '98')

```python
sequencia = input('Digite a seguencia de numeros: ')

lista_de_numeros = sequencia.split(',')
tupla_numeros = tuple(lista_de_numeros)
print(lista_de_numeros)
print(tupla_numeros)
```
